```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# x-grid
x = np.linspace(-2.0*np.pi, 2.0*np.pi, 500) 

#funtions to plot
y3 = x**3 
y2 = x**2 
y2x = 2*x 

#y-axis limit
plt.ylim(-6,6)

#plot lines
plt.plot(x, y3, lw=4, ls='-', label=r'$(x)^3$')
plt.plot(x, y2, lw=4, ls='--', label=r'$(x)^2$')
plt.plot(x, y2x, lw=4, ls='-', label=r'$2(x)$')

#display legend and define axises & plot title
plt.legend() 
plt.xlabel(r'$x$', fontsize=24) 
plt.ylabel(r'$y$', fontsize=24) 
plt.title('Functions Plot', fontsize=24)
plt.show()
```
